```python
import pandas as pd
import matplotlib
```

```python
#Read csv file and use built-in index
overdose_deaths = pd.read_csv('drug_deaths.csv', index_col=0)
overdose_deaths
```

```python
#Display first few rows of data set
overdose_deaths.head()
```

```python
#Describe dataset
overdose_deaths.describe()
```

```python
#Info on dataset
overdose_deaths.info()
```

```python
#Find missing values in dataset
missing_values = overdose_deaths.isnull()
print(missing_values)
```

```python
overdose_deaths
```

```python
#See columns in list
column_names_list = overdose_deaths.columns.tolist()
print(column_names_list)
#Drop ID, ResidenceCounty, ResidenceCity, DeathCity, Deathcounty, DeathCityGeo, ResidenceCityGeo, InjuryCityGeo, InjuryState, LocationifOther, InjuryCity, InjuryCounty, OtherSignifican, and ResidenceState to focus on Age, Sex, Race and the different Drugs
overdose_deaths_cleaned = overdose_deaths.drop(['ID', 'ResidenceCounty', 'ResidenceCity', 'InjuryState', 'DeathCity', 'DeathCounty', 'DeathCityGeo', 'ResidenceCityGeo', 'InjuryCityGeo', 'LocationifOther', 'InjuryCity', 'InjuryCounty', 'OtherSignifican', 'ResidenceState'], axis=1)
overdose_deaths_cleaned
```

```python
#See columns in list after dropping unwated columns
column_names_list = overdose_deaths_cleaned.columns.tolist()
print(column_names_list)
```

```python
#Look at data types
overdose_deaths_cleaned.dtypes
```

```python
#Convert Date column data type from object to datetime
overdose_deaths_cleaned['Date'] = pd.to_datetime(overdose_deaths_cleaned['Date']) 
overdose_deaths_cleaned.dtypes
```

```python
overdose_deaths_cleaned.value_counts()
```

```python
# Get the counts and percentages of males vs females
counts = overdose_deaths_cleaned['Sex'].value_counts(normalize=True)

# Convert to percentages
percentages = counts * 100

print(percentages)
```

```python
# Count the occurrences of each gender
gender_counts = overdose_deaths_cleaned['Sex'].value_counts()

# Plot the counts
gender_counts.plot(kind='bar')
```

```python
#Check null sums for each column
null_count = overdose_deaths_cleaned.isnull().sum()
print(null_count)
```

```python
overdose_deaths_cleaned.info()
```

```python
#Pull month from Date column

overdose_deaths_cleaned['month'] = overdose_deaths_cleaned['Date'].dt.month_name()

print(overdose_deaths_cleaned)
```

```python
#Pull year from Date column and convert Year data type

overdose_deaths_cleaned['year'] = pd.to_datetime(overdose_deaths_cleaned['Date'], format="%Y").dt.year


print(overdose_deaths_cleaned)
```

```python
#See columns in list after adding new month  and year column
column_names_list = overdose_deaths_cleaned.columns.tolist()
print(column_names_list)
```

```python
# Rename month column to Month and year to Year
overdose_deaths_cleaned = overdose_deaths_cleaned.rename(columns={'month': 'Month', 'year': 'Year'})
```

```python
#See columns in list after updating name of Month column
column_names_list = overdose_deaths_cleaned.columns.tolist()
print(column_names_list)
```

```python
#Check counts for Month column
overdose_deaths_cleaned.Month.value_counts()
```

```python
overdose_deaths_cleaned.dtypes
```

```python
#Check counts for Year column
overdose_deaths_cleaned.Year.value_counts()
```

```python
#Describe data after being cleaned
overdose_deaths_cleaned.describe()
```

```python
overdose_deaths_cleaned.dtypes
```

```python
#Check average age and min and max
avg_overdose_age = overdose_deaths_cleaned.Age.mean()
print(avg_overdose_age)

youngest_overdose_age = overdose_deaths_cleaned.Age.min()
print(youngest_overdose_age)

oldest_overdose_age = overdose_deaths_cleaned.Age.max()
print(oldest_overdose_age)

```

```python
# Count the occurrences of each Year
year_counts = overdose_deaths_cleaned['Year'].value_counts()

# Plot the counts
year_counts.plot(kind='bar')
```
